```python
import pandas as pd
import os
```

```python
CSV_PATH = os.path.join("..","..","source", "Sales_Dataset.csv")
if os.path.exists(CSV_PATH):
    df = pd.read_csv(CSV_PATH)
    print(df.head())
else:
    print(f"File not found: {CSV_PATH}")
```

```python
max_spent = df['Total Amount'].max()
print(f"Maximum amount spent in a single transaction: {max_spent}")
```

```python
min_spent = df['Total Amount'].min()
max_spent = df['Total Amount'].max()
print(f"Minimum amount spent in a single transaction: {min_spent}")
print(f"Maximum amount spent in a single transaction: {max_spent}")
```

```python
median_spent = df['Total Amount'].median()
print(f"Median amount spent in a single transaction: {median_spent}")
```

```python
import numpy as np

median_spent = df['Total Amount'].median()

# Fast vectorized way
df['spend_category'] = np.where(
    df['Total Amount'] > median_spent,
    'A',
    'B'
)

# Summary
print(df)
```
